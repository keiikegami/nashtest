```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
from scipy.misc import factorial as fact
```

```python
fact(4)
```

```python
#　思った感じ
a = np.linspace(-30, 0, 200)

def part(x, p):
    return ((-1)**p)*(2*p-1)*(x**(2*p+1))/(fact(2*p+1))

def func(x, order):
    return x + sum([part(x, i) for i in range(1,order)])

order = 100
result = [func(i, order) for i in a]
plt.plot(a, result)
plt.show()
```

```python
#　思ったより変な形だった
a = np.linspace(-10, 0, 200)

def part(x, p):
    return ((-1)**p)*(2*p-1)*(x**(2*p+1))/(fact(2*p+1))

def func(x, order):
    return x + sum([part(x, i) for i in range(1,order)])

order = 100
result = [func(i, order) for i in a]
plt.plot(a, result)
plt.xlabel(r"$\theta_{\delta}$")
plt.show()
plt.savefig("ident1.png")
```

```python
def cdf(x):
    return 1/2 + ((x**2) * np.cos(x) - 2*x*np.sin(x))/(np.sqrt(2*np.pi))
```

```python
a = np.linspace(-1, 1, 200)
result = [cdf(i) for i in a]
plt.plot(a, result)
plt.show()
```
