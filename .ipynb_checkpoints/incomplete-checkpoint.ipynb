```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
% matplotlib inline
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
e = st.norm()
```

```python
def br(mu,delta,p):
    return e.cdf(mu + delta*p)
```

```python
# 真のパラメータの時の均衡
xs = np.linspace(0,1,200)
mu = 0.5
delta = -1
result = [br(mu,delta,x) for x in xs]
plt.plot(xs, result)
plt.plot(result, xs)
plt.show()
```

```python
# muを変化させても単一の均衡
mus = np.linspace(-1,1,10)
delta = -1
for mu in mus:
    result = [br(mu,delta,x) for x in xs]
    plt.plot(xs, result)
    plt.plot(result, xs)
plt.show()
```

```python
# deltaを動かしても単一均衡
deltas = np.linspace(-2,0,10)
mu = 0.5
for delta in deltas:
    result = [br(mu,delta,x) for x in xs]
    plt.plot(xs, result)
    plt.plot(result, xs)
plt.show()
```
